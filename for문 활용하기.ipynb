```python
# Q1. 구구단

a = int(input())

for i in range(1, 10):
    print("{} * {} = {}".format(a, i, a*i))
```

```python
# Q2. A + B - 3

case = int(input())

for i in range(1, case+1):
    a, b = map(int, input().split())
    print("출력물은 {}".format(a+b))

```

```python
# Q3. Given n, sum all the integers below n 

sum = 0
n = int(input())

for i in range(1, n+1):
    sum = sum + i

print(sum)
```

```python
# Q6. 역순으로 N 찍기


# 이거 속도가 그렇게 빠르진 않음.
n = int(input())
result = []
for i in range(1, n+1):
    result.append(i)

result.reverse()


for element in result:
    print(element)
```

```python
# Q7. A+B 

T = int(input())

for i in range(1, T+1):
    a, b = map(int, input().split())
    print("Case #{}: {}".format(i,a+b))
```

```python
# Q8. A+B (tidy version)

T = int(input())

for i in range(1, T+1):
    a, b = map(int, input().split())
    print("Case #{}: {} + {} = {}".format(i,a,b,a+b))
```

```python
# Q9. 별찍기

n = int(input())

for i in range(1, n+1):
    print('*' * i)
```

```python
# Q10. 별찍기

n = int(input())

for i in range(1, n+1):
    print(' '*(n-i)+'*'*i)
```

```python
# Q11. 정수 X 보다 작은 elements

N, X = map(int, input().split())
my_list = list(map(int, input().split()))

li = []
for i in my_list:
    if i < X:
        li.append(i)
        
for element in li:
    print(element, end=' ')
```
