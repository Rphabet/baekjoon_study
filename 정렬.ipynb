```python
# Q1. 수 정렬. n개 주어졌을 때 오름차순으로 정렬

n = int(input())

i = 1
li = []
while i <= n:
    m = int(input())
    li.append(m)
    i += 1
    
li = list(set(li))
li.sort()

for item in li:
    print(item)
```
