```python
# Q1. 

n = int(input())

factorial = 1
for num in range(2, n+1):
    factorial = factorial * num
    
print(factorial)
```

```python
# Q1. v2 factorial 

def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
    
    
n = int(input())

print(factorial(n))
```

```python
# Q2. fibonacci

def fib(num):
    if num == 0:
        return 0
    elif num == 1 or num == 2:
        return 1
    else:
        return fib(num-2) + fib(num-1)
    
n = int(input())

print(fib(n))
```
