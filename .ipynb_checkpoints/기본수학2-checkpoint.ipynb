```python
# Q1. 소수찾기

N = int(input())
m = list(map(int, input().split()))

def prime(n):
    i = 1
    div = []
    while i <= n:
        if n % i == 0:
            div.append(i)
        i += 1
    if len(div) == 2:
        return True
    else:
        return False

count = 0
for element in m:
    count += int(prime(element))
    
print(count)


############# 효율적인 버전 

def prime_soln(n):
    num_set = set(range(2, n+1))    # 2부터 n+1까지 숫자 배열 만들기
    
    for i in range(2,n+1):
        if i in num_set:
            num_set -= set(range(i*2, n+1, i))
        
    answer = len(num_set)
    print(num_set)
    
    return answer

print(prime_soln(17))
```

```python
# Q2. 소수 찾기 + 소수의 합

def prime(n):
    i = 1
    div = []
    while i <= n:
        if n % i == 0:
            div.append(i)
        i += 1
    if len(div) == 2:
        return n
    else:
        return False

M = int(input())
N = int(input())

prime_list = []
for num in range(M,N+1):
    if prime(num) != False:
        prime_list.append(num)
if len(prime_list) == 0:
    print(-1)
else:
    print(sum(prime_list))
    print(min(prime_list))
```

```python
# Q3. 소인수분해

def factorisation(n):
    i = 2
    li = []
    while n != 1:
        if n % i == 0:
            n = n / i
            li.append(i)
        else:
            i += 1
    return li


n = int(input())
answer = factorisation(n)

for item in answer:
    print(item)
```

```python
# Q4. 두번째 방법 M 이상 N 이하 사이에 있는 소수 나열하기

m, n = map(int, input().split())

def prime_soln(n):
    num_set = set(range(2, n+1))    # 2부터 n+1까지 숫자 배열 만들기
    
    for i in range(2,n+1):
        if i in num_set:
            num_set -= set(range(i*2, n+1, i))   # i 배수 제거로 시간 효율 극대화

    
    return num_set

answer = list(prime_soln(n) - prime_soln(m-1)) 

for item in answer:
    print(item)
```
