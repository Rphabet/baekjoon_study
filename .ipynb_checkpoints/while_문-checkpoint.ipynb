```python
# Q1. A+B 

while True:
    a, b = map(int, input().split())
    if (a,b) == (0,0):
        break
    print(a+b)
```

```python
# Q2. A+B
# 다시 해보자

while True:
    try:
        a, b = map(int, input().split())
        print(a+b)
    except:
        break
        
        
# 테스트 케이스 개수가 정해지지 않았기 때문에
# try: 변수 A,B에 int형이 들어간다면, A+B의 값을 출력한다.
# except: try에 대한 에러가 발생한 경우(ex. a 1, 3, 2 ㄱ 입력)
# break: while문을 멈춘다.

```

```python
# Q3. 더하기 싸이클

# x 는 1의 자리
# y는 10의 자리
# z는 x+y

a = input()
num = a
if len(a) == 1:
    x = int(a)
    y = 0
else:
    x = int(a[0])
    y = int(a[1])


i = 1
while True:
    z = x + y
    z = str(z)
    if len(z) == 1:
        q = str(x) + z
    else:
        q = str(x) + z[1]
        
    if q == num:
        break
    else:
        if len(q) == 1:
            x = int(q)
            y = 0
        else:
            x = int(q[1])
            y = int(q[0])
        i += 1
        
print(i)
     
```

```python
# Q3. (다시) 더하기 싸이클

n = int(input())
origin = n
new = str(n)
i = 1

while True:
    if int(new) < 10:
        new = '0' + new
        x = new[1]
        z = x
        new = x + z 
    else:
        x = new[1]
        y = new[0]
        z = int(y) + int(x)
        if z < 10:
            new = x + str(z)
            new = str(int(new))
        else:
            new = x + str(z)[1]
        
    if int(new) == origin:
        break
        
    else:
        i += 1

print(i)
```
