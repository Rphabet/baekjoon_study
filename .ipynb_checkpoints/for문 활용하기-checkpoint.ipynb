```python
# Q1. 구구단

a = int(input())

for i in range(1, 10):
    print("{} * {} = {}".format(a, i, a*i))
```

```python
# Q2. A + B - 3

case = int(input())

for i in range(1, case+1):
    a, b = map(int, input().split())
    print("출력물은 {}".format(a+b))

```

```python
# Q3. Given n, sum all the integers below n 

sum = 0
n = int(input())

for i in range(1, n+1):
    sum = sum + i

print(sum)
```
