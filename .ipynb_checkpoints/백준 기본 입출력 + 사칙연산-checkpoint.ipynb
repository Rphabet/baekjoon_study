```python
# Q3. 두수 입력후 출력


## 방법 1
a, b = input("곱 할 두 수를 입력하세요: ").split()
print(a)
print(b)
print(type(a))
a = int(a)
print(type(a))
b = int(b)

result = a * b
print(result)


#####################

## 방법 2 map 사용하기!
a, b = map(int, input("두 수를 곱해보자").split())
print(a * b)
```

```python
# Q4. 두 수 입력후 나누기로 출력

a, b = map(float, input("두 수를 나눠보자").split())
print(a/b)
```

```python
# Q5. 사칙연산을 해보자

a, b = map(int, input().split())
print(a+b)
print(a-b)
print(a*b)
print(a//b)
print(a%b)
```

```python
# Q6. 나머지

a, b, c = map(int, input().split())

print((a + b) % c)
print(((a % c) + (b % c)) % c)
print((a * b) % c)
print((a % c) * (b % c) % c)
```

```python
# Q7. 곱의 법칙을 파헤쳐보자!

a = int(input())
b = int(input())

step1 = int(str(b)[2])

print(step1 * a)

step2 = int(str(b)[1])
print(step2 * a)

step3 = int(str(b)[0])
print(step3 * a)

result = a * b
print(result)
```
