```python
# Q4. 문자열 R 번 반복

def rep(n, string):
    li = []
    for char in string:
        li.append(char*n)
    new_str = li[0]
    for i in range(1,len(li)):
         new_str = new_str + li[i]
    
    return new_str

n = int(input())
i = 0

while i < n:
    S = input().split()
    num = int(S[0])
    print(rep(num, S[1]))
    i += 1
```

```python
# Q7. 상수

def reverse(word):
    n = len(word)
    new_word = word[n-1]
    for i in range(0,n-1):
        new_word += word[n-2-i]
    return new_word

a, b = input().split()

a_reverse = reverse(a)
b_reverse = reverse(b)

if int(a_reverse) > int(b_reverse):
    print(int(a_reverse))
else:
    print(int(b_reverse))
```

```python
# Q8. 전화 거는 시간

def alphabet_to_num(word):
    dic = {
        'a': 2,
        'b': 2,
        'c': 2,
        'd': 3,
        'e': 3,
        'f': 3,
        'g': 4,
        'h': 4,
        'i': 4,
        'j': 5,
        'k': 5,
        'l': 5,
        'm': 6,
        'n': 6,
        'o': 6,
        'p': 7,
        'q': 7,
        'r': 7,
        's': 7,
        't': 8,
        'u': 8,
        'v': 8,
        'w': 9,
        'x': 9,
        'y': 9,
        'z': 9
    }
    
    result = 0
    for char in word:
        result += dic[char.lower()] + 1
    
    return result

s = input()

print(alphabet_to_num(s))
```

```python
# Q9. 크로아티아 알파벳

def croc_word(word):
    dic = {'c=': 1,
           'c-': 1,
           'dz=': 1,
           'd-': 1,
           'lj': 1,
           'nj': 1,
           's=': 1,
           'z=': 1}
    

    count = 0
    n = len(word)
    for i in range(0,len(word)-1):
        new_word = word[i] + word[i+1]
        if new_word in dic:
            count += 1
    
    for i in range(0,len(word)-2):
        new_word = word[i] + word[i+1] + word[i+2]
        if new_word in dic:
#             print(new_word)
            count += 1
    
    n = n - count
    
    return n

# print(croc_word('dz=ak'))

s = input()

print(croc_word(s))
    
        
```
